## 计算几何中心方法

假设有n个点，按照顺时针排列

```python
def Geometric_center(X, Y):
    m = len(X)
    m -= 1
    A = 0
    cost_x = 0
    cost_y = 0
    for i in range(m):
        cost_x += (X[i] + X[i + 1]) * (X[i] * Y[i + 1] - X[i + 1] * Y[i])
        cost_y += (Y[i] + Y[i + 1]) * (X[i] * Y[i + 1] - X[i + 1] * Y[i])
    for i in range(m):
        A += 0.5 * (X[i] * Y[i + 1] - X[i + 1] * Y[i])
    x, y = (1 / (6 * A)) * cost_x, (1 / (6 * A)) * cost_y
    return x, y

X, Y = [0, 1, 2, 1, 0], [1, 0, 1, 2, 1]
x, y = Geometric_center(X, Y)
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y, c='b')
plt.scatter(x, y, c='r')
print(x, y)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 随机生成 n 个多边形顶点
n = 10  # 你可以根据需要改变顶点数
x_coords = np.random.uniform(0, 10, n)
y_coords = np.random.uniform(0, 10, n)

# 计算几何中心（质心）
centroid_x = np.mean(x_coords)
centroid_y = np.mean(y_coords)

x_list = x_coords + x_coords[0]
y_list = y_coords + y_coords[0]

f_x, f_y = Geometric_center(x_list, y_list)
# 可视化多边形及其几何中心
plt.figure()
plt.fill(x_coords, y_coords, alpha=0.3, edgecolor='black', label="Polygon")
plt.plot(centroid_x, centroid_y, 'ro', label="Centroid mean")
plt.scatter(f_x, f_y, c='y', label = 'Centroid s')
# 显示顶点和中心坐标
for i, (x, y) in enumerate(zip(x_coords, y_coords)):
    plt.text(x, y, f'({x:.2f}, {y:.2f})')

plt.text(centroid_x, centroid_y, f'Centroid mean\n({centroid_x:.2f}, {centroid_y:.2f})', color='red')
plt.text(f_x, f_y, f'Centroid s\n({f_x:.2f}, {f_y:.2f})', color='y')

plt.legend()
plt.show()

print(f"几何中心坐标mean: ({centroid_x:.2f}, {centroid_y:.2f})")
print(f"几何中心坐标s: ({f_x:.2f}, {f_y:.2f})")
```
